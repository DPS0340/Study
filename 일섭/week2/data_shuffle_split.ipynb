```python
import pandas as pd

df = pd.read_csv('./california_housing/data/housing_raw.csv')
```

```python
df.head()
```

```python
import numpy as np

df_shuffled = df.iloc[np.random.permutation(df.index)].reset_index(drop=True)
```

```python
df_shuffled.head()
```

```python
df_shuffled.to_csv('./california_housing/data/housing_shuffled.csv')
```

```python
df_shuffled.shape
```

```python
split_len = int(df_shuffled.shape[0]*0.8)

df_train = df_shuffled[:split_len]
df_test = df_shuffled[split_len:]
```

```python
df_train.to_csv('./california_housing/data/housing_shuffled_train.csv')
df_test.to_csv('./california_housing/data/housing_shuffled_test.csv')
```
